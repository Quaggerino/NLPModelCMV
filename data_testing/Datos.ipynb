```python
import pandas as pd

# Leer cada archivo CSV en un DataFrame
df_positivo = pd.read_csv('positivo.csv', encoding='utf-8')
df_negativo = pd.read_csv('negativo.csv', encoding='utf-8')
df_irrelevante = pd.read_csv('irrelevante.csv', encoding='utf-8')

# Añadir la columna 'target' a cada DataFrame
df_positivo['target'] = 2
df_negativo['target'] = 1
df_irrelevante['target'] = 0

# Concatenar los tres DataFrames en uno
df_total = pd.concat([df_positivo, df_negativo, df_irrelevante], ignore_index=True)

# (Opcional) Puedes mezclar los datos si quieres que estén distribuidos aleatoriamente
df_total = df_total.sample(frac=1).reset_index(drop=True)

# Mostrar las primeras filas del DataFrame resultante
df_total.head()

# Guardar el DataFrame combinado en un nuevo archivo CSV (opcional)
df_total.to_csv('combinado.csv', index=False, encoding='utf-8')
```

```python
import pandas as pd

def split_long_texts(df, max_length=1000):
    new_rows = []
    
    for index, row in df.iterrows():
        text = row['data']
        target = row['target']
        
        # Split texts longer than max_length
        while len(text) > max_length:
            split_index = text.rfind(' ', 0, max_length)  # Find the last space within max_length
            if split_index == -1:  # If there are no spaces to split at, force a split at max_length
                split_index = max_length
            
            new_rows.append({'data': text[:split_index], 'target': target})
            text = text[split_index+1:]  # Update text to the remaining part
            
        # Append the last part or the unsplit text
        new_rows.append({'data': text, 'target': target})
    
    return pd.DataFrame(new_rows)

# Usage example
df = pd.read_csv('combinado.csv')

df = split_long_texts(df)

# Optionally, display the first few rows to confirm it's loaded correctly.
df.head(10)
```
